Create a function that manages the range values in 'world_rank' columns. Replace the range with the average between the lower and the upper values

```python
def convert_string_to_number(column_value):
    
    if len(column_value) <= 3:
        return int(column_value)
    
    if len(column_value) == 7:
        pattern = re.compile("(\d+)\-(\d+)")
        occurrence = pattern.search(column_value)
        
        if occurrence:
            lower = int(occurrence.group(1))
            upper = int(occurrence.group(2))
            central_value = (upper + lower) / 2
            return int(central_value)
```

```python
def normalize(column_element):

  for university in school_and_country_name['university_name_normalized']:
    if column_element == university:
      return column_element

  name_column_element = re.compile(column_element)

  for university in school_and_country_name['university_name_normalized']:
    match = name_column_element.search(university)
    if match:
      return university
    else:
      name_university = re.compile(university)
      match_university = name_university.search(column_element)
      if match_university:
        return university
```

Define a function that take the rankings of each dataset, and returns the difference between the maximum and the minimum for each university

```python
def max_dif_rank(row):
    
    world_number_rank_shanghai = row['world_number_rank_shanghai']
    world_number_rank_times = row['world_number_rank_times']
    world_rank = row['world_rank']
    
    maximum_ranking = max(world_number_rank_shanghai,world_number_rank_times,world_rank)
    minimum_ranking = min(world_number_rank_shanghai,world_number_rank_times,world_rank)
    
    return (maximum_ranking - minimum_ranking)
```

Column 'num_students' has the character ',' for the thousands, resulting as an object type

```python
times_most_recent_for_max['num_students']
```

Replace the ',' character

```python
def convert_int(col):
    repl = int(col.replace(',', ''))
    
    return repl
```

Apply the function on 'num_students' column

```python
times_most_recent_for_max['num_students'] = times_most_recent_for_max['num_students'].apply(convert_int)
times_most_recent_for_max.head()
```

Create functions to calculate the number of females and males

```python
def compute_female(col):
    
    # Si trova il match con il group regex
    fem_regex = re.match('^(\d+) : (\d+)$', col['female_male_ratio'])
        
    # Si convertono in interi i valori
    female = round(int(fem_regex.group(1)) * col['num_students']/100)
    
    return female
    
def compute_male(col):
    
    # Si trova il match con il group regex
    male_regex = re.match('^(\d+) : (\d+)$', col['female_male_ratio'])
    
    # Si convertono in interi i valori
    male = round(int(male_regex.group(2)) * col['num_students']/100)
    
    return male
```

Apply the functions creating 'female' and 'male' columns

```python
times_most_recent_for_max['female'] = times_most_recent_for_max.apply(compute_female, axis=1)
times_most_recent_for_max['male'] = times_most_recent_for_max.apply(compute_male, axis=1)
```

Group by country, summing the number of females and males

Create a function that calculate the ratio females/males. If males or females are equal to 0, the function will return a value equal to 0, indicating an only males or only females university

```python
def ratio_fem_mal(row):
    if  (row['male'] == None) or (row['male'] == 0):
        return None
    else:
        return round(row['female'] / row['male']*100,1)
```
